```python
import os
import re

PATH = os.path.abspath(os.getcwd())
```

```python
def read_transcript(filename):
    with open(PATH + f"/transcripts/{filename}") as f:
        dialogue = f.read().strip()
    return dialogue


def write_transcript(dialogue, filename):
    with open(PATH + f"/transcripts/{filename}", "w") as f:
        f.write(dialogue)


def swap_actors(dialogue, old_actor, new_actor):
    pattern = r"\[" + re.escape(old_actor) + r"\]:"
    replacement = f"[{new_actor}]:"
    return re.sub(pattern, replacement, dialogue)
```

```python
scenes =[['amadeus', 'constanze', ["Mozart", "Constanze", "Schikaneder"]],
         ['bound', 'coffee', ["Violet", "Corky"]],
         ['cedar_rapids', 'car', ["Tim", "Ronald", "Dean", "Joan"]],
         ['troy', 'achilles', ["Achilles", "Odysseus"]]]

base_dialogue = read_transcript("ex_machina.txt")

# Evaluations
results = []
for (game, scene, actors) in scenes:
    dialogue = swap_actors(base_dialogue, "Nathan", actors[0])
    dialogue = swap_actors(dialogue, "Caleb", actors[1])

    print(dialogue)
    print()
    print()

    write_transcript(dialogue, game + "/transcript_baseline.txt")
```
